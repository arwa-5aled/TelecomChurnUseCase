```python
def plot_cat_vs_target(df, feature, target='Churn'):
  
    ctab = pd.crosstab(df[feature], df[target], normalize='index') * 100
    display(ctab.round(2))   # show churn % table
    
  
    plt.figure(figsize=(6,4))
    sns.countplot(x=feature, hue=target, data=df, palette="Set2")
    plt.title(f"{feature} vs {target}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
df.info()
```

```python
categorical_features = [
    'gender','Senior_Citizen ','Is_Married','Dependents',
    'Phone_Service','Internet_Service','Online_Security','Online_Backup',
    'Device_Protection','Tech_Support','Streaming_TV','Streaming_Movies',
    'Contract','Paperless_Billing','Payment_Method'
]
for feature in categorical_features:
    plot_cat_vs_target(df, feature)
```

```python
def plot_churn_rate(df, feature, target='Churn'):
    churn_rate = df.groupby(feature)[target].apply(lambda x: (x=='Yes').mean()*100)
    churn_rate.plot(kind='bar', figsize=(6,4), color="tomato")
    plt.ylabel("Churn Rate (%)")
    plt.title(f"Churn Rate by {feature}")
    plt.xticks(rotation=45)
    plt.show()

for col in categorical_features:
    plot_churn_rate(df, col)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

num_features = ['tenure', 'Monthly_Charges', 'Total_Charges']

for col in num_features:
    plt.figure(figsize=(6,4))
    sns.histplot(data=df, x=col, hue='Churn', kde=True, bins=30)
    plt.title(f"Distribution of {col} by Churn")
    plt.show()
```

```python
for col in num_features:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='Churn', y=col, data=df, palette="Set2")
    plt.title(f"{col} vs Churn")
    plt.show()
```

```python
for col in ['tenure', 'Monthly_Charges', 'Total_Charges']:
    plt.figure(figsize=(6,4))
    sns.violinplot(x='Churn', y=col, data=df, split=True, inner="quartile")
    plt.title(f"{col} vs Churn")
    plt.show()
```

```python
for col in ['tenure','Monthly_Charges','Total_Charges']:
    plt.figure(figsize=(6,4))
    sns.kdeplot(df[df['Churn']=="Yes"][col], label="Churn=Yes", shade=True)
    sns.kdeplot(df[df['Churn']=="No"][col], label="Churn=No", shade=True)
    plt.title(f"Density of {col} by Churn")
    plt.legend()
    plt.show()
```
