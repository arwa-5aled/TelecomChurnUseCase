```python
def show_unique_values(df):
    for col in df.columns:
        unique_vals = df[col].unique()
        print(f"\nColumn: {col}")
        print(f"Number of unique values: {len(unique_vals)}")
        print(f"Unique values: {unique_vals[:15]}")  # show only first 15 to avoid clutter
```

```python
show_unique_values(df)
```

```python
df['Total_Charges'] = df['Total_Charges'].fillna(0)
```

```python
mask = (df['Dual'] == 'No phone service')
invalid_rows = df[mask & (df['Phone_Service'] != 'No')]
print(f"Number of inconsistent rows: {len(invalid_rows)}")
display(invalid_rows[['customerID', 'Phone_Service', 'Dual']])
```

```python
internet_related = [
    'Online_Security','Online_Backup','Device_Protection',
    'Tech_Support','Streaming_TV','Streaming_Movies'
]

mask = df['Internet_Service'] == 'No'

for col in internet_related:
    inconsistent = df[mask & (df[col] != 'No internet service')]
    print(f"{col}: {len(inconsistent)} inconsistent rows")
    if not inconsistent.empty:
        display(inconsistent[['CustomerId','Internet_Service',col]])
```

```python
mask = (df['Dual'] == 'No phone service')
invalid_rows = df[mask & (df['Phone_Service'] != 'No')]
print(f"Number of inconsistent rows: {len(invalid_rows)}")
display(invalid_rows[['customerID', 'Phone_Service', 'Dual']])
```
